# Installing Python Libraries with pip: Example with matplotlib

This notebook explains how to use `pip` to install Python libraries, using `matplotlib` as an example.

## 1. What is pip?
`pip` is the package installer for Python. You can use it to install packages from the Python Package Index (PyPI) and other indexes.

## 2. Installing matplotlib
To install `matplotlib`, you can run the following command in a code cell or in your terminal (without the ! if you're using a terminal):



```python
!pip install matplotlib
```

## 3. Importing and Using matplotlib
Once installed, you can import and use `matplotlib` in your Python code. Here is a simple example that creates a line plot:

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Simple Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

---
If you see an error that `matplotlib` is not found, make sure you have run the pip install command above.
